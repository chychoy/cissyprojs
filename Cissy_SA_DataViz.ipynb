### Bootstrapping Functions

```python
def bootstrap_mean(data, label, replications):
    """NOTE: all empty responses(5) are removed prior to
    calculation.
    label: label of column containing the variable
    replications: number of bootstrap samples
    Returns array of bootstrap sample means
    """
    means = np.array([])
    data = data[data[label]!=0]
    for i in np.arange(replications):
        bootstrap_sample = data.sample(len(data),replace=True)
        resampled_mean = np.mean(bootstrap_sample[label])
        means = np.append(means, resampled_mean)
    return means
    
```

```python
#bootstrap_mean(sa_data, "Sanctions for misconduct ", 1)
```

```python
def ci95(means):
    """Calculates the 95th confidence interval of the
    bootstrapped means from the given sample, returns
    lower and upper bounds as a list"""
    lower = np.percentile(means, 2.5)
    upper = np.percentile(means, 97.5)
    return lower,upper
```

```python
ci95(bootstrap_mean(sa_data, "Sanctions for misconduct", 100))
```

```python
def no_null_mean(data,label):
    """Calculates the mean of the given column label
    after removing the no-response selection (5)"""
    data = data[data[label]!= 0]
    data_mean = np.mean(data[label])
    return data_mean
```

### Statistical Confidence Interval

n = ~43 (because there are empty responses)
sample mean = see below

sample mean +- 1.96 (sample sd)/sqrt(n)

```python
def no_null_std(data,label):
    """Calculates the sample std of the given column label
    after removing the no-response selection (5)"""
    data = data[data[label]!= 0]
    data_std = np.std(data[label])
    return data_std
```

```python
def samp_n(data, label):
    """Returns the sample size after removing the empty values"""
    data = data[data[label]!= 0]
    data_n = len(data[label])
    return data_n
```

```python
samp_mean = np.array([])
for heading in headings_array:
    samp_mean = np.append(samp_mean, no_null_mean(sa_data, heading))
```

```python
samp_std = np.array([])
for heading in headings_array:
    samp_std = np.append(samp_std, no_null_std(sa_data, heading))
```

```python
samp_ns = np.array([])
for heading in headings_array:
    samp_ns = np.append(samp_ns, samp_n(sa_data, heading))
```

```python
values2 = pd.DataFrame({"titles":headings_array,
                        "n": samp_ns, 
                        "std": samp_std,
                       "mean": samp_mean})
```

```python
values2 = values2.sort_values(by = ["mean"], ascending = True).reset_index(drop = True)
```

```python
values2.head()
```

```python
values2["lower"] = values2["mean"] - 1.96*(values2["std"]/(values2["n"])**(1/2))
values2["upper"] = values2["mean"] + 1.96*(values2["std"]/(values2["n"])**(1/2))
```
